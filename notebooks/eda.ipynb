```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plus = pd.read_csv("../data/A_plus_normalized_levels.csv")
minus = pd.read_csv("../data/A_minus_normalized_levels.csv")
split = pd.read_csv("../data/split_to_train_validation_test_disjoint_sets_ids.csv")
seq = pd.read_csv("../data/mRNA_sequences.csv")
split
```

```python
plus.columns = ["id"] + [f"{i}h+" for i in range(1, 9)] + ["10h+"]
minus.columns = ["id"] + [f"{i}h-" for i in range(1, 9)] + ["10h-"]
```

```python
df = pd.merge(seq, plus, on="id")
df = pd.merge(df, minus, on="id")
df
```

```python
train_df = df.iloc[split["train_ids"]]
val_df = df.iloc[split["validation_ids"].dropna().map(int)]
test_df = df.iloc[split["test_ids"].dropna().map(int)]
```

```python
train_df.to_csv("../data/train.csv", index=False)
val_df.to_csv("../data/val.csv", index=False)
test_df.to_csv("../data/test.csv", index=False)
```

```python
df = pd.read_csv("../data/test.csv")
test_mac = df[df["1h+"] >= 2.0]
test_mac.to_csv("../data/test_mac.csv", index=False)
```
